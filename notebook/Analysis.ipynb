# Analysis of the experiments done with the Bayesian optimisation

Here, we will plot the data of experiments with the Bayesian optimisation.

There are four types of graphics:
- the exploration in the sensorial space (the number of differents cases explored) trough the iteration, the curve is mean of several iterations and the shaded area is plus and minus the standard deviation, the parameter epsilon represents the coefficient of exploration and p the number of iterations on the same goal;
- the number of balls catched by the arm (in the **ArmBall** environment);
- the boxplot of the time of computation;
- the measure of competence, the distances to random sensorial goals of the answer of 4 methodes in the same database, the Nearest Neighbor and the Bayesian optimisation for 1 and 20 iterations on the same goal, the parameter b represent the number of iteration on the same goal.
